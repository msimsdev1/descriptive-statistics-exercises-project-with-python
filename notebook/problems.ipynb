# Descriptive statistics problems

### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Set random seed for reproducibility
np.random.seed(42)

# Make distributions - both normal and Chi-Square
normal_dist = np.random.normal(loc=0, scale=1, size=100)
chi_square_dist = np.random.chisquare(df=3, size=100)

# Statistical metrics function
def calculate_metrics(data):
    return {
        'Mean': np.mean(data),
        'Median': np.median(data),
        'Standard Deviation': np.std(data),
        'Variance': np.var(data),
        'Skewness': stats.skew(data),
        'Kurtosis': stats.kurtosis(data),
        'Minimum': np.min(data),
        'Maximum': np.max(data),
        'Range': np.ptp(data),
        '25th Percentile': np.percentile(data, 25),
        '50th Percentile': np.percentile(data, 50),
        '75th Percentile': np.percentile(data, 75),
        '100th Percentile': np.percentile(data, 100)
    }

# Print metrics
print("Normal Distribution Metrics:")
for metric, value in calculate_metrics(normal_dist).items():
    print(f"{metric}: {value:.4f}")

print("\nChi-Square Distribution Metrics:")
for metric, value in calculate_metrics(chi_square_dist).items():
    print(f"{metric}: {value:.4f}")

# Visualization
plt.figure(figsize=(12, 5))

# Normal Distribution Histogram - in blue
plt.subplot(1, 2, 1)
plt.hist(normal_dist, bins=10, edgecolor='black')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Chi-Square Distribution Histogram - in orange
plt.subplot(1, 2, 2)
plt.hist(chi_square_dist, bins=10, color='orange', edgecolor='black')
plt.title('Chi-Square Distribution (df=3)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Modify layout and display
plt.tight_layout()
plt.show()

# Additional comprehensive statistical analysis
print("\n--- Additional Statistical Analysis ---")

# Normality Test (Shapiro-Wilk Test)
print("\nNormality Test (Shapiro-Wilk Test):")
normal_shapiro = stats.shapiro(normal_dist)
chi_square_shapiro = stats.shapiro(chi_square_dist)
print(f"Normal Distribution - Statistic: {normal_shapiro.statistic:.4f}, p-value: {normal_shapiro.pvalue:.4f}")
print(f"Chi-Square Distribution - Statistic: {chi_square_shapiro.statistic:.4f}, p-value: {chi_square_shapiro.pvalue:.4f}")

# Descriptive Statistics
print("\nDescriptive Statistics:")
print("Normal Distribution:")
print(f"Confidence Interval (95%): {stats.t.interval(confidence=0.95, df=len(normal_dist)-1, loc=np.mean(normal_dist), scale=stats.sem(normal_dist))}")
print("Chi-Square Distribution:")
print(f"Confidence Interval (95%): {stats.t.interval(confidence=0.95, df=len(chi_square_dist)-1, loc=np.mean(chi_square_dist), scale=stats.sem(chi_square_dist))}")

# Other Calculations
print("\nMoment Calculations:")
print("Normal Distribution:")
print(f"1st Moment (Mean): {stats.moment(normal_dist, moment=1):.4f}")
print(f"2nd Moment (Variance): {stats.moment(normal_dist, moment=2):.4f}")
print(f"3rd Moment (Skewness): {stats.moment(normal_dist, moment=3):.4f}")
print(f"4th Moment (Kurtosis): {stats.moment(normal_dist, moment=4):.4f}")
print("Chi-Square Distribution:")
print(f"1st Moment (Mean): {stats.moment(chi_square_dist, moment=1):.4f}")
print(f"2nd Moment (Variance): {stats.moment(chi_square_dist, moment=2):.4f}")
print(f"3rd Moment (Skewness): {stats.moment(chi_square_dist, moment=3):.4f}")
print(f"4th Moment (Kurtosis): {stats.moment(chi_square_dist, moment=4):.4f}")

# Quartile and Interquartile Range Analysis
print("\nQuartile Analysis:")
print("Normal Distribution:")
print(f"Q1 (25th Percentile): {np.percentile(normal_dist, 25):.4f}")
print(f"Q2 (50th Percentile - Median): {np.percentile(normal_dist, 50):.4f}")
print(f"Q3 (75th Percentile): {np.percentile(normal_dist, 75):.4f}")
print(f"Q3 (100th Percentile): {np.percentile(normal_dist, 100):.4f}")
print(f"Interquartile Range (IQR): {np.percentile(normal_dist, 75) - np.percentile(normal_dist, 25):.4f}")
print("Chi-Square Distribution:")
print(f"Q1 (25th Percentile): {np.percentile(chi_square_dist, 25):.4f}")
print(f"Q2 (50th Percentaile Median): {np.percentile(chi_square_dist, 50):.4f}")
print(f"Q3 (75th Percentile): {np.percentile(chi_square_dist, 75):.4f}")
print(f"Q4 (100th Percentile): {np.percentile(chi_square_dist, 100):.4f}")
print(f"Interquartile Range (IQR): {np.percentile(chi_square_dist, 75) - np.percentile(chi_square_dist, 25):.4f}")
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
import numpy as np
import matplotlib.pyplot as plt

# Define our sample dataset
data = [4, 2, 5, 8, 6]

# (SD)Standard Deviation using NumPy
std_dev = np.std(data)

# Show Mean
mean = np.mean(data)

# Variance and (MSD) Manual Standard Deviation
variance = sum((x - mean) ** 2 for x in data) / len(data)
manual_std_dev = variance ** 0.5

# Print results
print(f"Data: {data}")
print(f"Standard Deviation (NumPy): {std_dev}")
print(f"\nMean: {mean}")
print(f"Variance: {variance}")
print(f"Manual Standard Deviation: {manual_std_dev}")

# Plot Creation
plt.figure(figsize=(10, 6))

# Plot the data points
plt.scatter(range(len(data)), data, color='blue', label='Data Points')

# Plot mean line
plt.axhline(y=mean, color='red', linestyle='--', label='Mean')

# Plot the standard deviation range
plt.fill_between(range(len(data)), mean - std_dev, mean + std_dev, 
                 color='gray', alpha=0.3, label='1 Std Dev Range')

# Customize the plot
plt.title('Data Points with Mean and Standard Deviation')
plt.xlabel('Data Point Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)

# Plot of data
plt.show()
````
