### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Create distributions (both normal and Chi-Square)
normal_dist = np.random.normal(loc=0, scale=1, size=100)
chi_square_dist = np.random.chisquare(df=3, size=100)

# Create a figure with two subplots to differentiate between the two distribution
plt.figure(figsize=(12, 5))

# Normal Distribution Histogram - in blue
plt.subplot(1, 2, 1)
plt.hist(normal_dist, bins=10, edgecolor='black')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Chi-Square Distribution Histogram -in orange
plt.subplot(1, 2, 2)
plt.hist(chi_square_dist, bins=10, color='orange', edgecolor='black')
plt.title('Chi-Square Distribution (df=3)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Modify layout and display
plt.tight_layout()
plt.show()
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
import numpy as np

# Define our sample dataset
# This represents a small set of measurements or values we want to analyze
data = [4, 2, 5, 8, 6]

# Use NumPy to calculate standard deviation
# Calculate the population standard deviation
std_dev = np.std(data)

# Print out the original data and its standard deviation for verification
print(f"Data: {data}")
print(f"Standard Deviation (NumPy): {std_dev}")

# Calculate the mean (average) of the dataset
mean = np.mean(data)

# Calculate variance: average of squared differences from the mean
# Variance measures how spread out the numbers are from their average
variance = sum((x - mean) ** 2 for x in data) / len(data)

# Standard deviation is the square root of variance
manual_std_dev = variance ** 0.5

# Print the manually calculated standard deviation for comparison
print(f"\nMean: {mean}")
print(f"Variance: {variance}")
print(f"Manual Standard Deviation: {manual_std_dev}")
````
